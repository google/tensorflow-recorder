# Loading a TensorFlow Dataset

This notebook briefly demonstrates how to load a TF Dataset from TFRecord files generated by TFRecorder.
Note that currently, the TFRecord files must be in a directory on your local machine.
The directory is expected to have the following structure, based on TFRecorder's `convert` default output:
```
tfrecord_dir/
    train-*.tfrecord.gz
    validation-*.tfrecords.gz
    test-*.tfrecord.gz
    transformed_metadata/
    transform_fn/
```

The last two TF Transform directories are necessary in order to parse the TFRecord files properly.

````python
import tfrecorder
````

We'll use the sample TFRecord directory under `tfrecorder/test_data`.

````python
tfrecord_dir = '../tfrecorder/test_data/sample_tfrecords/'

datasets = tfrecorder.load(tfrecord_dir)
````

The `dataset.load` function returns a `dict` of `split`, `dataset` pairs.

````python
datasets.keys()
````

Let's examine the contains of the training set

````python
train = datasets['TRAIN']
````
